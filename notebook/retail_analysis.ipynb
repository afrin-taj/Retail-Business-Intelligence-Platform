```python
#checking for n
print(df.isnull().sum())
```

```python
df['Item Weight'] = df['Item Weight'].fillna(df['Item Weight'].median())
```

```python
# Standardize fat content
df['Item Fat Content'] = df['Item Fat Content'].replace({
    'LF': 'Low Fat',
    'low fat': 'Low Fat',
    'reg': 'Regular'
})
```

```python
df['Item Fat Content'].unique()
```

```python
df.describe()
```

```python
#Category generating more revenue
category_sales = (
    df.groupby('Item Type')['Sales']
      .sum()
      .sort_values(ascending=False)
)

category_sales.head(10)
```

```python
#outlet types generate the most revenue
outlet_sales = (
    df.groupby('Outlet Type')['Sales']
      .sum()
      .sort_values(ascending=False)
)

outlet_sales
```

```python
#location tier performance
location_sales = (
    df.groupby('Outlet Location Type')['Sales']
      .sum()
      .sort_values(ascending=False)
)

location_sales
```
